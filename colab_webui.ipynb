```
# 파일 내용 보기
file_path = "/content/GPT-SoVITS/tools/asr/fasterwhisper_asr.py"

with open(file_path, "r") as f:
    lines = f.readlines()

# transcribe가 있는 줄 확인 및 수정
for i, line in enumerate(lines):
    if "model.transcribe" in line and "language" not in line:
        lines[i] = line.replace("model.transcribe(", 'model.transcribe(language="ko", task="transcribe", ')

# 수정 후 저장
with open(file_path, "w") as f:
    f.writelines(lines)

print("✅ 수정 완료: transcribe에 language='ko' 적용됨")
```
